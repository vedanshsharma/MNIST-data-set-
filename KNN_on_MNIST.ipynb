```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics import roc_auc_score
import seaborn as sns
```

```
#loading the data
data= pd.read_csv("/content/drive/My Drive/MNIST/train.csv")
l=data['label']
d=data.drop('label',axis=1)
print(d.shape)
print(l.shape)
del(data)#free wasted memory
```

```
#splitting into train and test set
#splitting into train test and cross validation set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(d, l, test_size=0.25, random_state=0)
print("shape of training data ", X_train.shape)
print("shape of test data ", X_test.shape)
#freeing useless memory
del(d)
del(l)
```

#KNN

```
from sklearn.neighbors import KNeighborsClassifier
import scikitplot.metrics as skplt
def KNN(X_train,y_train,X_test,y_test, algo):
  clf = KNeighborsClassifier(n_neighbors=102, algorithm= algo)
  clf.fit(X_train,y_train)
  probs= clf.predict_proba(X_test)
  print("AUC score on test data", roc_auc_score(y_test,probs,multi_class='ovr'))
  prediction = clf.predict(X_test)
  skplt.plot_confusion_matrix(y_test,prediction)
  print("macro f1 score for data :",metrics.f1_score(y_test, prediction, average = 'macro'))
  print("micro f1 scoore for data:",metrics.f1_score(y_test, prediction, average = 'micro'))
  print("hamming loss for data:",metrics.hamming_loss(y_test,prediction))
  print("Precision recall report for data:\n",metrics.classification_report(y_test, prediction))
```

```
#Brute force
KNN(X_train,y_train,X_test,y_test, 'brute')
```

```
#kd tree
KNN(X_train,y_train,X_test,y_test, 'kd_tree')
```

```
#ball tree
KNN(X_train,y_train,X_test,y_test, 'ball_tree')
```
